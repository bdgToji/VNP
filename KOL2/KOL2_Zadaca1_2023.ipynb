# Направете претпроцесирање на податоците

```python
df
```

```python
df.isnull().sum()
```

```python
lag = 2
```

```python
for i in range(1, lag+1):
  df[f'USDT_prev_{i}'] = df['Close (USDT)'].shift(i)
```

```python
df = df.dropna()
```

```python
df = df.set_index('Date')
```

```python
df
```

# Поделето го на тренирачко и тестирачко множество во сооднос 80:20

```python
from sklearn.model_selection import train_test_split
```

```python
x, y = df.drop(columns=['Close (USDT)']), df['Close (USDT)']
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, shuffle=False)
```

```python
X_train.head()
```

```python
X_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Искористете соодветен Boosting модел за проблемот

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(learning_rate=0.6,max_depth=15,alpha=5,n_estimators=35)
model.fit(X_train, Y_train)
```

```python
pred = model.predict(X_test)
```

# Тестирајте го моделот на тестирачкото множество со метриките за евалуација на дадениот проблем

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(Y_test, pred)
```

```python
r2_score(Y_test, pred)
```
